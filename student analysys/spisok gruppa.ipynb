Импортируем библиотеки

```python
import json
import statistics
import matplotlib.pyplot as plt
```

Открываем файлик, создаем словарь

```python
with open ("data.json", 'r') as data:
    d = json.load(data)
```

Считаем средний рост, вес и отклонения

```python
sur = []
suv = []
imya = []
for i in range (1, 23):
    imya.append(d[str(i)]['Name'])
    sur.append(d[str(i)]['Rost'])
    suv.append(d[str(i)]['Ves'])   
print('Средний рост: ', statistics.mean(sur), 'Отклонение: ', statistics.pstdev(sur))
print('Средний вес: ', statistics.mean(suv), 'Отклонение: ', statistics.pstdev(suv))
```

Строим график роста

```python
plt.plot(imya, sur)
print('График роста')
plt.show()
```

Строим график веса

```python
plt.plot(imya, suv)
print('График веса')
plt.show()
```

Сортируем словарь по росту

```python
sur_2 = sorted(sur)
index = 1
porostu = {}
for i in sur_2 :
    for j in d.values():
        if  j['Rost'] == i:
            porostu[str(index)] = j
            index+= 1
print(porostu)
print(type(porostu))
```

Записываем отсортированный словарь в файл

```python
with open('porostu.json', 'w') as f:
    json.dump(porostu,f)
```

Строим график отсортированного словаря

```python
porostugr = []
for i in range (1, 23):
    porostugr.append(porostu[str(i)]['Rost'])
plt.plot(imya, porostugr)
print('График роста')
plt.show()
```
